```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import ndimage
```

```python
p = 0.3
M = 1000
L = 50
plt.figure(figsize = [3.2, 4.8])
allarea = np.array([])
for j in range(M):
    z = np.random.rand(L,L,L)
    m = z<p
    lw, num = ndimage.label(m)
    labelList = np.arange(lw.max() + 1)
    area = ndimage.sum(m, lw, labelList)
    allarea = np.append(allarea, area)
n, sbins = np.histogram(allarea, 1000)
s = 0.5*(sbins[1:]+sbins[:-1])
nsp = n/(L*L*L*M)
k = np.nonzero(n)
plt.plot(np.log(s[k]), np.log(nsp[k]), '.', c = "mediumaquamarine")
plt.xlabel('$log(s)$', fontsize = 15)
plt.ylabel('$log(n(s,p))$', fontsize = 15)
tit = 'p='+str(p)
plt.title(tit, fontsize = 15)
coeffs = np.polyfit(np.log(s[k][10:150]), np.log(nsp[k][10:150]), 1)
poly = coeffs[1] + coeffs[0]*np.log(s[k][10:150])
tau = -(poly[-1]-poly[0])/(np.log(s[k])[150]-np.log(s[k])[10])
tau = round(tau, 2)
plt.plot(np.log(s[k][10:150]), poly, linestyle = "-", linewidth = 2.5, c="black")
leg  = 'Ï„='+str(tau)
plt.legend(['$log(n(s,p))$', 'Line of Best Fit, ' + leg], fontsize=10)
```
